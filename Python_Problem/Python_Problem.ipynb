```python
from twython import Twython
import jsonlines
```

```python

CONSUMER_KEY = "< Consumer Key >"
CONSUMER_SECRET = "< Consumer Secret >"
ACCESS_KEY = "< Access Key >"
ACCESS_SECRET = "< Access Secret >" 
```

```python
# Function to get the tweets
def get_tweets(username):
    
    # Pass consumer key, consumer secret,access key and access secret to Twython
    tweeter = Twython(CONSUMER_KEY,CONSUMER_SECRET,ACCESS_KEY,ACCESS_SECRET)
    
    # Get tweets as list
    tweets = tweeter.get_user_timeline(screen_name = username,count = 200)
    
    
    return tweets
    
```

```python
# Function to save tweets into jsonlines file
def tweets_to_jsonlines(tweets,file_name):
    with jsonlines.open(file_name,mode="w") as writer:
        for tweet in tweets:
            writer.write(tweet)
    return file_name
```

```python
# Function to display tweets from jsonlines file
def display_tweets_from_jsonlines(file_name):
    with jsonlines.open(file_name,mode="r") as reader:
        for i,json_obj in enumerate(reader):
            print("-----------------------------------------------------------------------------------------------------")
            print("    Tweet " + str(i+1) + ":\n-----------------------------------------------------------------------------------------------------")
            print("Date and time: " + json_obj['created_at']+"\n")
            print("Text: " + json_obj['text'] + "\n")
            print("No. of favorites: " + str(json_obj['favorite_count'])+"\n")
            print("No. of retweets: " + str(json_obj['retweet_count'])+"\n")
            if("media" in json_obj['entities']):
                print("No. of images: " + str(len(json_obj['entities']['media'])))
            else:
                print("No. of images: None")
                      
            print("-----------------------------------------------------------------------------------------------------\n\n\n")
```

```python
if __name__ == '__main__': 
  
    # whose tweets are to be extracted. 
    file_name = "Tweets.jsonl"
    tweets = get_tweets("midasIIITD") 
    tweets_to_jsonlines(tweets,file_name)
    display_tweets_from_jsonlines(file_name)
```
